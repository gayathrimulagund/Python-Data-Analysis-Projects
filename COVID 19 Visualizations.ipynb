## Max infection rate per country

```python
confirmed.head()
```

```python
countries = list(df['Country'].unique())
countries
```

```python
# max per country
max_infection_rates = []
for c in countries:
    MIR = confirmed[confirmed.Country == c].Confirmed.diff().max()
    max_infection_rates.append(MIR)
print(max_infection_rates)
```

```python
df_MIR = pd.DataFrame()
df_MIR['Country'] = countries
df_MIR['Max Infection Rate'] = max_infection_rates
df_MIR.head()
```

```python
# PLot these values
df_MIR, x='Country',y='Max Infection Rate',color='Country',title='Global Max Infection Rate',log_y=True)
```

## Lockdown impacts in Italy

```python
italy_lockdown_start_date = '2020-03-09'
italy_lockdown_a_month_later = '2020-04-09'
```

```python
confirmed.head()
```

```python
Italy.head()
```

```python
Italy['Infection Rate'] = Italy.Confirmed.diff()
Italy.head()
```

```python
fig2 = px.line(Italy, x='Date',y='Infection Rate', title='Before and After Lockdown in Italy')
fig2.show()
```

```python
fig2 = px.line(Italy, x='Date',y='Infection Rate', title='Before and After Lockdown in Italy')
fig2.add_shape(
    dict(
    type='line',
    x0=italy_lockdown_start_date,
    y0=0,
    x1=italy_lockdown_start_date,
    y1=Italy['Infection Rate'].max(),
    line = dict(color='red', width=2)
)
)
fig2.add_annotation(
    dict(
    x=italy_lockdown_start_date,
    y=Italy['Infection Rate'].max(),
    text="starting date of lockdown"
    )
)

fig2.add_shape(
    dict(
    type='line',
    x0=italy_lockdown_a_month_later,
    y0=0,
    x1=italy_lockdown_a_month_later,
    y1=Italy['Infection Rate'].max(),
    line = dict(color='orange', width=2)
)
)
fig2.add_annotation(
    dict(
    x=italy_lockdown_a_month_later,
    y=Italy['Infection Rate'].max(),
    text = "ending date of lockdown"
    )
)
```

## Lockdown iimpact on Deaths in Italy

```python
# Deaths rate claculations
Italy['Deaths Rate'] = Italy.Deaths.diff()
Italy.head()
```

```python
# Visualization
fig3 = px.line(Italy,x='Date', y=['Infection Rate', 'Deaths Rate'])
fig3.show()
```

```python
# Normalizing the columns

Italy['Infection Rate'] = Italy['Infection Rate']/Italy['Infection Rate'].max()
Italy['Deaths Rate'] = Italy['Deaths Rate']/Italy['Deaths Rate'].max()
```

```python
# Visulization
fig4 = px.line(Italy, x='Date', y=['Infection Rate', 'Deaths Rate'])
fig4.show()
```
